# 머신러닝 준비(차트 그리기)

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
n= np.random.randint(0,10,10)
a= np.random.randint(0,10,10)
index = range(10)
plt.bar(index,n)
plt.plot(index,a,'r')
#r을 ro 로 바꾸면 산점도로 나옴
plt.show()
```

```python
x=np.arange(-5, 5, 0.1)
y=x**2

plt.plot(x,y,'r')
plt.show()
```

```python
x=np.arange(-5, 5, 0.1)
a=0.1
b=0

y=a*x+b

plt.plot(x,y,'r')
plt.show()
```

```python
x=np.arange(-5, 5, 0.1)
a=3
b=6
y=a*x+b
u=np.arange(-5, 5, 0.1)
a=0.1
b=12
c=a*x+b

v=np.arange(-5, 5, 0.1)
a=5
b=3
z=a*x+b


plt.plot(x,y,'r')
plt.plot(u,c,'b')
plt.plot(v,z,'g')
plt.legend(y)
plt.legend(c)
plt.legend(z)
plt.show()
```
